```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import fbeta_score
from sklearn.metrics import recall_score
import numpy as np
# 실제 레이블
actual_labels = [0, 1, 0, 1, 1, 2, 0, 2, 1, 2]

# 예측된 레이블
predicted_labels = [0, 1, 1, 1, 0, 2, 0, 2, 2, 2]

# confusion matrix 계산
cm = confusion_matrix(actual_labels, predicted_labels)

# confusion matrix를 비율로 변환하여 계산
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

# seaborn을 사용하여 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(cm_normalized, annot=True, fmt=".1%", cmap="Blues")

# 개수와 백분율 함께 표시
# for i in range(cm.shape[0]):
#     for j in range(cm.shape[1]):
        # text = f"{cm[i, j]} ({cm_normalized[i, j] * 100:.1f}%)"
        # text = f"{cm[i, j]} ({cm_normalized[i, j]})"
        # plt.text(j + 0.5, i + 0.4, text, ha='center', va='center')

plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("Actual Labels")
plt.show()

p = precision_score(actual_labels, predicted_labels, average='macro')

print(p)
print(classification_report(actual_labels, predicted_labels))
print(accuracy_score(actual_labels, predicted_labels))
print(precision_recall_fscore_support(actual_labels, predicted_labels))
print("f1_score: ", fbeta_score(actual_labels, predicted_labels, beta=1,average='macro'))
print("recall_score: ", recall_score(actual_labels, predicted_labels, average='macro'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

# 실제 레이블
y_true = [0, 1, 0, 1, 1, 2, 0, 2, 1, 2]

# 예측된 레이블
y_pred = [0, 1, 1, 1, 0, 2, 0, 2, 2, 2]


# 실제 레이블과 예측된 레이블로 confusion matrix 생성
cm = confusion_matrix(y_true, y_pred)

# 정확도 계산
accuracy = accuracy_score(y_true, y_pred)

# F1 스코어 계산
f1 = f1_score(y_true, y_pred, average='weighted')

# 클래스 레이블
labels = ['class 1', 'class 2', 'class 3']

# 시각화
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('prediction')
plt.ylabel('GT')
plt.title('Confusion Matrix')
plt.text(0.5, -0.15, f'Accuracy: {accuracy:.2f}, F1 Score: {f1:.2f}', ha='center', transform=plt.gca().transAxes)
plt.show()
```
